# Visualization of Missing Value Location
This notebook explains visualization techniques for the location of missing values using the followings:
 1. seaborn.heatmap function
   - Just-one-line, easy, and instant method.
 2. [missingno](https://github.com/ResidentMario/missingno) module
   - Requires an additional module, but gives us more insight on missing values.

## Library import, setup, and data preparation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sns.set()
```

```python
df = pd.read_csv('train.csv') # Use Titanic train dataset from Kaggle as an example
```

```python
df.head()
```

```python
# df.info() shows the number of non-null values (and hence the number of missing values).
df.info()
```

## 1. Using seaborn.heatmap function

```python
# Following command shows missing/non-missing values in two colors.
sns.heatmap(df.isnull(), cbar=False)
```

## 2. Using [missingno](https://github.com/ResidentMario/missingno) module

```python
import missingno as msno
```

```python
# missingno.matrix function shows missing/non-missing values in two colors.
# The result is similar to seaborn.heatmap function above, 
# plus the bar on the right side which indicates completeness of values in each entry 
# (ranges from 10 to 12 in this dataset.)
msno.matrix(df)
```

```python
# missingno.heatmap function gives nullity correlation
# (how strongly the presence or absence of one variable affects the presence of another).
# For Titanic train dataset, there is no strong correlation observed for nullity.
msno.heatmap(df)
```
